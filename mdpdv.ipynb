```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load datasets (you can adjust file paths as necessary)
heart_df = pd.read_excel('heart.xlsx')
diabetes_df = pd.read_excel('diabetes.xlsx')
parkinsons_df = pd.read_excel('parkinsons.xlsx')
```

```python
heart_df.head()
```

```python
# @title cp

from matplotlib import pyplot as plt
heart_df['cp'].plot(kind='hist', bins=20, title='cp')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
diabetes_df.head()
```

```python
# @title Glucose vs BloodPressure

from matplotlib import pyplot as plt
diabetes_df.plot(kind='scatter', x='Glucose', y='BloodPressure', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
parkinsons_df.head()
```

```python

# ------- Heart Disease Visualizations -------
# Pie chart for Heart Disease distribution
heart_counts = heart_df['target'].value_counts()
plt.pie(heart_counts, labels=['No Disease', 'Has Disease'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'salmon'])
plt.title('Heart Disease Distribution')
plt.show()
```

```python

# Bar chart for Age vs. Heart Disease
sns.histplot(data=heart_df, x='age', hue='target', multiple='stack', palette='coolwarm')
plt.title('Age Distribution vs Heart Disease')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python

# Create a crosstab of sex vs. target
crosstab = pd.crosstab(heart_df['sex'], heart_df['target'])

# Plot the stacked bar chart
crosstab.plot(kind='bar', stacked=True, color=['red', 'lightblue'])
plt.title('Gender vs Heart Disease (Stacked)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'], rotation=0)
plt.legend(title='Heart Disease')
plt.show()
```

```python

# ------- Diabetes Visualizations -------
# Pie chart for Diabetes distribution
diabetes_counts = diabetes_df['Outcome'].value_counts()
plt.pie(diabetes_counts, labels=['No Diabetes', 'Has Diabetes'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'orange'])
plt.title('Diabetes Distribution')
plt.show()
```

```python

# Histogram of Glucose levels
sns.histplot(data=diabetes_df, x='Glucose', hue='Outcome', kde=True, palette='coolwarm')
plt.title('Glucose Levels Distribution')
plt.xlabel('Glucose Level')
plt.ylabel('Density')
plt.show()
```

```python

# Scatter plot for BMI vs. Glucose
sns.scatterplot(x='BMI', y='Glucose', hue='Outcome', data=diabetes_df, palette='coolwarm')
plt.title('BMI vs Glucose Levels by Diabetes Outcome')
plt.xlabel('BMI')
plt.ylabel('Glucose Level')
plt.show()
```

```python

# ------- Parkinsons Disease Visualizations -------
# Pie chart for Parkinson's Disease distribution
parkinsons_counts = parkinsons_df['status'].value_counts()
plt.pie(parkinsons_counts, labels=['No Parkinsons', 'Has Parkinsons'], autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue'])
plt.title('Parkinson\'s Disease Distribution')
plt.show()
```

```python

# Box plot for voice features (Jitter and Shimmer)
plt.figure(figsize=(10, 6))
sns.boxplot(x='status', y='MDVP:Jitter(%)', data=parkinsons_df)
plt.title('Jitter Percentage Comparison')
plt.xlabel('Parkinson\'s Status')
plt.ylabel('Jitter (%)')
plt.show()
```

```python

# Scatter plot for Jitter vs. Shimmer
sns.scatterplot(x='MDVP:Jitter(%)', y='MDVP:Shimmer', hue='status', data=parkinsons_df, palette='coolwarm')
plt.title('Jitter vs Shimmer by Parkinson\'s Status')
plt.xlabel('Jitter (%)')
plt.ylabel('Shimmer')
plt.show()
```
